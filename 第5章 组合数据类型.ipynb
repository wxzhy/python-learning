### 集合
集合元素之间无序，每个元素唯一，不存在相同元素  
集合用大括号{} 表示，元素间用逗号分隔  
建立集合类型用{} 或set()  
集合元素不可更改，不能是可变数据类型  
建立空集合类型，必须使用set()

```python
A = {'Python', 'Lina', 'Gong'}
print(A)
```

```python
B = set("PYPY1122")
print(B)
```

```python
B = set()
print(type(B))
b = {}
print(type(b))
```

```python
S = {"Python", "Lina", "Gong"}
T = {"Python", "Suchun", "Ma"}
S |= T
print(S)
S = {"Python", "Lina", "Gong"}
T = {"Python", "Suchun", "Ma"}
S -= T
print(S)
S = {"Python", "Lina", "Gong"}
T = {"Python", "Suchun", "Ma"}
S &= T
print(S)
S = {"Python", "Lina", "Gong"}
T = {"Python", "Suchun", "Ma"}
S ^= T
print(S)
```

```python
S = {"Lina", "Gong"}
print(S)
S.add("Ma")
S.add("Suchun")
print(S)
#移除元素，元素不在集合中，不报错
S.discard("Li")
print(S)
#元素不在集合中，会报错
# S.remove("Zhao")
print(S)
```

```python
S = {"Lina", "Gong"}
S.add("Ma")
S.add("Suchun")
S.add("Zixi")
for item in S:
    print(item, end=" ")
```

包含关系比较

```python
print("Gong" in {"Python", "Gong"})
print({"Li"} >= {"Gong"})
```

数据去重：集合类型所有元素无重复

```python
ls = {"Python", "Lina", "Gong", "Ma", "Suchun", "Ma", "Zixi"}
S = set(ls)
print(S)
```

### 元组
元组是序列类型的一种扩展，一旦创建就不能被修改  
使用小括号()或tuple()创建，元素间用,分隔

```python
t1 = tuple()
t2 = tuple("abc")
t3 = tuple([1, 1, 1])
t4 = tuple(range(3))
print(t1, t2, t3, t4)
```

可以使用或不使用小括号

```python
creature = "cat", "dog", "tiger", "human"
print(creature[::-1])
color = ("blue", creature)
print(color[-1][2])  #读取倒数第一个元素
```

### 列表
列表是一种序列类型，创建后可以随意被修改  
使用方括号[]或list()创建，元素间用逗号，分隔  
列表中各元素类型可以不同，无长度限制


```python
ls = ["cat", "dog", "tiger"]
print(ls)
lt = ls
print(lt)
```

```python
ls = ["cat", "dog", "tiger"]
ls[1:2] = [1, 2, 3]
print(ls)
del ls[::3]
print(ls)
print(ls * 2)
```

```python
ls = ["dragon", "cat", "dog"]
ls.append("tiger")
print(ls)
ls.insert(4, "human")  #插入
ls.insert(5, "胜太")
print(ls)
```

如果不希望数据被程序所改变，转换成元组类型

```python
ls = ["dragon", "cat", "dog", "Tiger", "Human"]
print(ls)
lt = tuple(ls)
print(lt)
```

### 字典
每个键对应于一个值。在字典中，键不能重复。  
字典的键只能使用不可变的对象，但字典的值可以使用不可变或可变的对象  
字典是键值对的集合，键值对之间无序  
采用大括号{}和dict()创建，键值对用冒号: 表示。
[ ] 用来向字典变量中索引或增加元素。


```python
d = {"中国": "北京", "美国": "华盛顿", "德国": "柏林"}
print(d)
print(d["中国"])
d["法国"] = "巴黎"
print(d)
print("印度" in d)
print(d.keys())
print(d.values())
print(d.items())
```

```python
d = {"中国": "北京", "美国": "华盛顿", "德国": "柏林"}
capital = d.get("中国")
print(capital)
capital = d.get("加拿大", "渥太华")
print(capital)
print(d.popitem())
```
