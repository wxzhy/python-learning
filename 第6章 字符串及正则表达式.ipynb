输入任意字符串，统计其中元音字母（‘a’、‘e’、'i'、'o'、'u'，不区分大小写）出现的次数和频率

```python
s1=input("请输入字符串：")
s=s1.upper()
lentotal=len(s)
counta=s.count('A')
counte=s.count('E')
counti=s.count('I')
counto=s.count('O')
countu=s.count('U')
print('所有字母的总数为：',lentotal)
print('元音字母出现的次数和频率分别为：')
print('A: {0}\t{1:2.2f}%'.format(counta,counta/lentotal*100))
print('E: {0}\t{1:2.2f}%'.format(counte,counte/lentotal*100))
print('I: {0}\t{1:2.2f}%'.format(counti,counti/lentotal*100))
print('O: {0}\t{1:2.2f}%'.format(counto,counto/lentotal*100))
print('U: {0}\t{1:2.2f}%'.format(countu,countu/lentotal*100))
```

### 正则表达式
匹配单个字符

| **元字符** | **含义**                                         |
| ---------- | ------------------------------------------------ |
| .          | 匹配单个字符,DOTALL模式中可以匹配'\n'            |
| \d         | 数字:[0-9]                                       |
| \D         | 非数字:[^\d]                                     |
| \w         | 匹配任意一个普通字符  [_0-9a-zA-Z]               |
| \W         | 匹配任意的非普通字符[^_0-9a-zA-Z],即匹配特殊字符 |
| \s         | 匹配任意空字符                                   |
| \S         | 匹配任意非空字符                                 |
| []         | 匹配括号内的任意'一个'字符                       |
| [^]        | 除了字符集中的任意'一个'字符                     |

匹配重复性

| **元字符** | **含义**                |
| ---------- | ----------------------- |
| *          | 重复前面的字符0次或多次 |
| +          | 重复前面的字符1次或多次 |
| ?          | 重复前面的字符0次或1次  |
| {n}        | 重复前面的字符n次       |

匹配位置

| **元字符** | **含义**                                     |
| ---------- | -------------------------------------------- |
| ^          | 匹配字符串开头。在多行模式中匹配每一行的开头 |

转义
    
| **元字符** | **含义**                           |

re常用的功能函数
Compile：编译正则表达式模式，返回一个对象的模式。可以把那些常用的正则表达式编译成正则表达式对象，以提高效率。

格式：
	re.compile(pattern,flags=0)
	pattern: 编译时用的表达式字符串
	flags:编译标志位用于修改正则表达式的匹配方式，如：是否区分大小写，多行匹配等。常用的flags有：
	

| **标志**             | **含义**                        |
| -------------------- | ------------------------------- |
| re.S   re.DOTALL     | 使.匹配包括\n在内的所有字符     |
| re.I   re.IGNORECASE | 使匹配对大小写不敏感            |
| re.M  re.MULTILINE   | 元字符^ $可以匹配每行的开头结尾 |
| re.X   re.VERBOSE    | 可以给正则添加注释              |

**使用按位或'|'连接多个flags**


```python
import re
pattern=re.compile(r'\d+')
m=pattern.match('one12twothree34four')
print(m)
m=pattern.match('one12twothree34four',2,10)
print(m)
m=pattern.match('one12twothree34four',3,10)
print(m)
print(m.group())
```

**match object**对象的常用方法：

   **group(n = 0)**

​     功能：获取match对象匹配到的内容

​     参数：默认获取正则整体匹配到的内容，如果赋值1,2,3...表示获取某个子组匹配到的内容

​     返回值:返回匹配内容

   **groups** 

​     功能：将所有子组匹配内容形成一个元组

```python
import re
line="Cats are smarter than dogs"
matchObj=re.match(r'(.*) are (.*?) .*',line,re.M|re.I)
if matchObj:
    print("matchObj.groups():",matchObj.groups())
    print("matchObj.group():",matchObj.group())
    print("matchObj.group(1):",matchObj.group(1))
    print("matchObj.group(2):",matchObj.group(2))
else:
    print("No match!!")
```

re.match()与search()区别  

re.match只匹配字符串的开始，如果字符串开始不符合正则表达式，则匹配失败，函数返回None；而re.search匹配整个字符串，直到找到一个匹配。

```python
import re
line="Cats are smarter than dogs"
searchObj=re.search(r'(.*) are (.*?) .*',line,re.M|re.I)
if searchObj:
    print("searchObj.groups():",searchObj.groups())
    print("searchObj.group():",searchObj.group())
    print("searchObj.group(1):",searchObj.group(1))
    print("searchObj.group(2):",searchObj.group(2))
else:
    print("Nothing found!!")
```

```python
import re
line="Cats are smarter than dogs"
matchObj=re.match(r'dogs',line,re.M|re.I)
if matchObj:
    print("match --> matchObj.group():",matchObj.group())
else:
    print("No match!!")
searchObj=re.search(r'dogs',line,re.M|re.I)
if searchObj:
    print("search --> searchObj.group():",searchObj.group())
else:
    print("Nothing found!!")
```

**注意**： match 和 search 是匹配一次 findall 匹配所有。


```python
import re
pattern=re.compile(r'\d+') # 查找数字
result1=pattern.findall('runoob 123 google 456')
result2=pattern.findall('run88oob123google456',0,10)
print(result1)
print(result2)
```

finditer和 findall 类似，在字符串中找到正则表达式所匹配的所有子串，并把它们作为一个迭代器返回。

```python
import re
it=re.finditer(r'\d+',"12a32bc43jf3")
for match in it:
    print(match.group())
```

### jieba库的使用
精确模式：就是把一段文本精确地切分成若干个中文单词，若干个中文单词之间经过组合，就精确地还原为之前的文本。其中**不存在冗余**单词。

全模式：将一段文本中所有可能的词语都扫描出来，可能有一段文本它可以切分成不同的模式，或者有不同的角度来切分变成不同的词语，在全模式下，Jieba库会将各种不同的组合都挖掘出来。分词后的信息再组合起来**会有冗余**，不再是原来的文本。


搜索模式：在精确模式基础上，对发现的那些长的词语，对它再次切分，进而适合搜索引擎对短词语的索引和搜索，也有冗余。


主要函数：


```python
import jieba
txt=open("测试文件.txt",'r',encoding='utf-8').read()
words=jieba.lcut(txt) #精确模式
counts={} #通过字典存储词语和出现次数
for word in words:
    if len(word)==1: #排除单个字符的分词结果
        continue
    else:
        counts[word]=counts.get(word,0)+1
items=list(counts.items()) #转换成列表
items.sort(key=lambda x:x[1],reverse=True) #按出现次数从大到小排序
for i in range(15):
    word,count=items[i]
    print("{0:<5}{1:>5}".format(word,count))
```
