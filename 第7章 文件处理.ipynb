### 文件操作
默认打开字符集为gbk?

```python
#文本形式打开文件
tf=open('f.txt','rt')
print(tf.readline())
tf.close()
```

二进制模式打开为binary类型

```python
#二进制形式打开文件
bf=open('f.txt','rb')
print(bf.readline())
bf.close()
```

Windows下文件名不区分大小写，但是Linux下区分大小写  
Windows下可用\\或/，Linux下只能用/


文件打开后可用用file.close()主动关闭

```python
fname=input('请输入要打开的文件名称:')
fo=open(fname,'r')
txt=fo.read()
#进行处理
print(txt)
fo.close()
```

### 按数量读入  
因为内存有限，若文件较大，read()和readlines()的一次读取便会耗尽系统内存。
为了保证读取安全，通常多次调用read()方法，每次读取size字节的数据。


```python
fname=input('请输入要打开的文件名称:')
fo=open(fname,'r')
txt=fo.read(2)
while txt!='':
    #进行处理
    print(txt)
    txt=fo.read(2)
fo.close()
```

### 文件的内容的读取

```python
fname=input('请输入要打开的文件名称:')
fo=open(fname,'r')
for line in fo.readlines():
    print(line)
fo.close()
```

```python
fname=input('请输入要打开的文件名称:')
fo=open(fname,'r')
for line in fo:
    print(line)
fo.close()
```

### 上下文管理语句with
能够保证文件正常关闭

```python
with open('test.txt', 'r') as src, open('test_new.txt', 'w') as dst:
	dst.write(src.read())
```

### 文件目录操作
使用OS模块

```python
import os
print(os.getcwd())#获取当前工作目录
os.chdir(r'..')#切换当前目录
print(os.listdir(r'.'))#列举当前目录中的内容
os.makedirs(r'temp/dir1')#创建多级子目录。注：mkdir()只能创建一级子目录
os.rename(r'temp/dir1',r'temp/dir2')#子目录重命名
```
