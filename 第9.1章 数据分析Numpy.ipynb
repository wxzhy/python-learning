 数组重塑 ：对于定义好的数组，可以通过reshape方法改变其数据维度。  
                         `np.reshape(a, newshape, order='C')`

| **参数名称** | **说明**                 |
| ------------ | ------------------------ |
| a            | 需要处理的数据。         |
| newshape     | 新维度——整数或整数元组， |


```python
arr1=np.arange(8)
print(arr1)
arr1.reshape(4,2)
print(arr1)
```

 reshape的参数中的其中一个可以设置为-1，表示数组的维度可以通过数据本身来推断。

```python
arr1=np.arange(12)
print('arr1:',arr1)
arr2=arr1.reshape(2,-1)
print('arr2:',arr2)
```

与reshape相反的方法是数据散开（ravel）
或数据扁平化（flatten）。

```python
arr1=np.arange(12).reshape(3,4)
print('arr1:',arr1)
arr2=arr1.ravel()
print('arr2:',arr2)
```

### 数组合并
hstack函数：实现横向合并
vstack函数：实现纵向组合是利用vstack将数组纵向合并
concatenate函数：可以实现数组的横向或纵向合并，参数axis=1时进行横向合并，axis=0时进行纵向合并


```python
import numpy as np
if __name__ == '__main__':
    arr1=np.arange(6).reshape(3,2)
    print(arr1)
    arr2=arr1*2
    print(arr2)
    arr3=np.hstack((arr1,arr2))
    print(arr3)
```

### 数组分隔
与数组合并相反，hsplit函数、vsplit函数和split函数分别实现数组的横向、纵向和指定方向的分割。


```python
import numpy as np
if __name__ == '__main__':
    arr=np.arange(16).reshape(4,4)
    print(arr)
    print('横向分割为:\n',np.hsplit(arr,2))
    print('纵向分割为:\n',np.vsplit(arr,2))
```

### 数组转置和轴对换
数组转置是数组重塑的一种特殊形式，可以通过transpose方法进行转置。

### 数组索引和切片
一维数组的索引类似Python中的列表。  
数组的切片返回的是原始数组的视图，不会产生新的数据，如果需要的并非视图而是要复制数据，则可以通过copy方法实现



```python
arr=np.arange(10)
print(arr)
print(arr[2])
print(arr[-1])
print(arr[1:4])
```

多维数组索引和切片：对于多维数组，它的每一个维度都有一个索引，各个维度的索引之间用逗号分隔。

```python
import numpy as np
if __name__ == '__main__':
    arr=np.arange(12).reshape(3,4)
    print(arr)
    print(arr[0,1:3]) #索引第一行的第二、三列
    print(arr[:,2]) #索引第三列
    print(arr[:1,:1]) #索引第一行第一列
```

### Numpy中数据分析与统计
Sort函数：对数据直接进行排序，调用改变原始数组，无返回值。  
`numpy.sort(a, axis, kind, order) `  
  
np.argsort函数和np.lexsort函数根据一个或多个键值对数据集进行排序。  
np.argsort():  返回的是数组值从小到大的索引值;  
np.lexsort():  返回值是按照最后一个传入数据排序的结果。  



```python
import numpy as np
if __name__ == '__main__':
    arr=np.array([7,9,5,2,9,4,3,1,4,3])
    print('原数组：',arr)
    arr.sort()
    print('排序后：',arr)
    a=[1,5,7,2,3,-2,4]
    b=[9,5,2,0,2,8,7]
    ind=np.lexsort((b,a))
    print('ind:',ind)
    tmp=[(a[i],b[i]) for i in ind]
    print('tmp:',tmp)
    ind=np.lexsort((a,b))
    print('ind:',ind)
    tmp=[(a[i],b[i]) for i in ind]
    print('tmp:',tmp)
```

### Numpy中数据分析与统计
重复数据与去重
在NumPy中，对于一维数组或者列表，unique函数去除其中重复的元素，并按元素由大到小返回一个新的元组或者列表。

```python
import numpy as np
if __name__ == '__main__':
    names=np.array(['红色','蓝色','黄色','红色','白色','红色'])
    print('原数组：',names)
    print('去重后的数组：',np.unique(names))
```

统计分析中有时也需要把一个数据重复若干次，使用tile和repeat函数即可实现此功能。


```python
import numpy as np
if __name__ == '__main__':
    arr=np.arange(5)
    print('原数组：',arr)
    wy=np.tile(arr,3)
    print('重复数据处理：\n',wy)
```
